```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from pandas_datareader import data, wb
import datetime
```

```python
#Define Start/End dates, and ticker names
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

```python
# Amazon
AMZN = data.DataReader("AMZN", 'yahoo', start, end)
AMZN
```

```python
myAMZN = AMZN.reset_index()
```

```python
# Plot a time series (line chart) of the “volume” from January 1, 2022 to today
plt.style.use('bmh')
myAMZN.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

```python
# Plot a bar chart of the “opening“ for the month of December 2021
newstart = datetime.datetime(2021, 12, 1)
newend = datetime.datetime(2021, 12, 31)
newAMZN = data.DataReader("AMZN", 'yahoo', newstart, newend)
newAMZN
```

```python
myNewAMZN = newAMZN.reset_index()
```

```python
myNewAMZN.plot.bar(x = 'Date', y = 'Open')
```

```python
# Microsoft
MSFT = data.DataReader("MSFT", 'yahoo', start, end)
MSFT
```

```python
myMSFT = MSFT.reset_index()
```

```python
# Plot a time series (line chart) of the “volume” from January 1, 2022 to today
plt.style.use('bmh')
myMSFT.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```
